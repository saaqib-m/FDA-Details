```python
import requests
import pandas as pd 
import time
```

```python
# use requests and json to obtain the detailws from the api
response = requests.get('https://api.fda.gov/food/enforcement.json?limit=10').json()
```

```python
# the 'results' section obtains the necessary information and [0] gets the first entry
response['results'][0]
```

```python
# prduce df for data storage
df = pd.DataFrame(columns=["Country","City","Recall Reason","Quantity","Product Description","Class","Firm","Type","Status"])
```

```python
# function to obtain fda details 

def get_fda_details(df):

    for vid in response['results']:
        country = vid['country']
        city = vid['city']
        recall_reason = vid['reason_for_recall']
        quantity = vid['product_quantity']
        product_desc = vid['product_description']
        classi = vid['classification']
        firm = vid['recalling_firm']
        types = vid['product_type']
        status = vid['status']
        
        df = df.append({"Country":country,"City":city,"Recall Reason":recall_reason,"Quantity":quantity,
                        "Product Description": product_desc,"Class":classi,"Firm":firm,"Type":types,"Status":status}, ignore_index=True
        )

    return df
```

```python
df = get_fda_details(df)
```
